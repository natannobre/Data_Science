```python
import pandas as pd

# Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt
```

```python
avocados = pd.read_csv("datasets/avocado.csv")
```

```python
# Look at the first few rows of data
print(avocados.head())
```

```python
# Get the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby("type")["AveragePrice"].sum()

# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind="bar", rot=45)

# Show the plot
plt.show()
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby("Date")["AveragePrice"].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind="line")

# Show the plot
plt.show()
```

```python
# Scatter plot of nb_sold vs avg_price with title
avocados.plot(x="Total Volume", y="AveragePrice", kind="scatter", title="Volume vs. average price")

# Show the plot
plt.show()
```

```python
# Histogram of conventional avg_price 
avocados[avocados["type"] == "conventional"]["AveragePrice"].hist()

# Histogram of organic avg_price
avocados[avocados["type"] == "organic"]["AveragePrice"].hist()

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

```python
# Modify histogram transparency to 0.5 
avocados[avocados["type"] == "conventional"]["AveragePrice"].hist(alpha=0.5)

# Modify histogram transparency to 0.5
avocados[avocados["type"] == "organic"]["AveragePrice"].hist(alpha=0.5)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

```python
# Modify bins to 20
avocados[avocados["type"] == "conventional"]["AveragePrice"].hist(alpha=0.5, bins=20)

# Modify bins to 20
avocados[avocados["type"] == "organic"]["AveragePrice"].hist(alpha=0.5, bins=20)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

```python
# Check individual values for missing values
print(avocados.isna())

# Check each column for missing values
print(avocados.isna().any())

# Bar plot of missing values by variable
avocados.isna().sum().plot(kind="bar")

# Show plot
plt.show()
```

```python
# Remove rows with missing values
avocados_complete = avocados.dropna()

# Check if any columns contain missing values
print(avocados_complete.isna().any())
```

```python
# List the columns with missing values
cols_with_missing = ["Small Bags", "Large Bags", "XLarge Bags"]

# Create histograms showing the distributions cols_with_missing
avocados[cols_with_missing].hist()

# Show the plot
plt.show()
```
